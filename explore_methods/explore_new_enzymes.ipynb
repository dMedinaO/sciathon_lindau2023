```python
import pandas as pd

new_enzymes = pd.read_csv("../data_to_sciathon_new/glove.csv")
smiles_used_to_train = pd.read_csv("../results/coding_smiles/mol2vec_smiles.csv")
```

```python
new_enzymes
```

```python
#concat
matrix_data = []
matrix_with_data = []

for index in new_enzymes.index:
    print("Processing index: ", index)
    for index2 in smiles_used_to_train.index:
        row = []
        
        for column in new_enzymes.columns:
            if "p_" in column:
                row.append(new_enzymes[column][index])
        for column in smiles_used_to_train.columns:
            if "p_" in column:
                row.append(smiles_used_to_train[column][index2])
        
        row_data = [new_enzymes['id_seq'][index], new_enzymes['type'][index], smiles_used_to_train['SMILE'][index2]]
        matrix_data.append(row)
        matrix_with_data.append(row_data)

header = ["p_{}".format(i) for i in range(len(matrix_data[0]))]
df_data = pd.DataFrame(matrix_data, columns=header)

df_data_info = pd.DataFrame(matrix_with_data, columns=['id_seq', 'type', 'smile'])
```

```python
df_data_info
```

```python
from joblib import load

model_trained = load("../results/training_model_exploration/model_optim_with_balanced_class.joblib")
```

```python
predictions = model_trained.predict(df_data)
df_predictions = pd.DataFrame()
df_predictions['prediction'] = predictions

df_predictions_filters = df_predictions.loc[df_predictions['prediction'] == 1]
len(df_predictions_filters)
```

```python
df_dta_filter = df_data_info.filter(items = df_predictions_filters.index, axis=0)
df_dta_filter.smile.unique()
```
