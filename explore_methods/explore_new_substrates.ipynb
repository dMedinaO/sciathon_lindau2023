```python
import pandas as pd

new_smiles = pd.read_csv("../data_to_sciathon_new/process_new_smiles.csv")
glove_enzymes = pd.read_csv("../results/coding_protein/glove.csv")
```

```python
glove_enzymes
```

```python
#concat
matrix_data = []

for index in new_smiles.index:
    for index2 in glove_enzymes.index:
        row = []
        for column in glove_enzymes.columns:
            if "p_" in column:
                row.append(glove_enzymes[column][index2])
        for column in new_smiles.columns:
            if "p_" in column:
                row.append(new_smiles[column][index])
        
        matrix_data.append(row)

header = ["p_{}".format(i) for i in range(len(matrix_data[0]))]
df_data = pd.DataFrame(matrix_data, columns=header)
```

```python
from joblib import load

model_trained = load("../results/training_model_exploration/model_optim_with_all_examples.joblib")
```

```python
predictions = model_trained.predict(df_data)
predictions
```
