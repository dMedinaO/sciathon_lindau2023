```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-white")
plt.rc('axes', grid=False, facecolor="white")
plt.rcParams.update({'font.size': 14})
```

```python
df_data = pd.read_csv("../../results/training_model_exploration/data_group_by.csv")
df_data
```

```python
df_data.encoding_protein.unique()
```

```python
filter_Data = ['albert', 'bepler', 'bert', 'esm1b', 'fasttext', 'glove', 'onehot', 'plus_rnn', 'seq_vec', 't5_xl', 'word2vec', 'xlnet']
df_data_filter_NLP = df_data.loc[df_data['encoding_protein'].isin(filter_Data)]

df_data_filter_NLP['Encoding Method'] = 'Natural Language Processing'
```

```python
filter_Data = ['group0', 'group1', 'group2', 'group3', 'group4','group5', 'group6', 'group7', ]
df_data_filter_p = df_data.loc[df_data['encoding_protein'].isin(filter_Data)]

df_data_filter_p['Encoding Method'] = 'Physicochemical properties'
```

```python
filter_Data = ['group0_FFT', 'group1_FFT', 'group2_FFT', 'group3_FFT', 'group4_FFT','group5_FFT', 'group6_FFT', 'group7_FFT' ]
df_data_filter_FFT = df_data.loc[df_data['encoding_protein'].isin(filter_Data)]

df_data_filter_FFT['Encoding Method'] = 'Frequency space'
```

```python
data_concat = pd.concat([df_data_filter_NLP, df_data_filter_p, df_data_filter_FFT], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.boxplot(data=data_concat, x="Encoding Method", hue="Algorithm", y="Accuracy", orient="v")

plt.savefig("data_fig_1.svg")
```
